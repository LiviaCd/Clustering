```python
import numpy as np
def calculate_elbow_method(data, max_k=10):
    inertia = []
    k_values = range(1, max_k + 1)
    
    for k in k_values:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(data)
        inertia.append(kmeans.inertia_)
    
    plt.figure(figsize=(10, 6))
    plt.plot(k_values, inertia, 'bo-')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Inertia (Sum of Squared Distances)')
    plt.title('Elbow Method for Optimal k')
    plt.grid(True)
    
    second_derivative = np.diff(np.diff(inertia))
    elbow_index = np.argmax(second_derivative) + 2  # +2 because of double diff and 1-indexed
    optimal_k = k_values[elbow_index]
    
    plt.axvline(x=optimal_k, color='r', linestyle='--', 
                label=f'Suggested optimal k = {optimal_k}')
    plt.legend()
    plt.show()
    
    return optimal_k, inertia

# Run the elbow method calculation on your data
optimal_k, inertia_values = calculate_elbow_method(data)
print(f"Suggested optimal number of clusters based on Elbow Method: {optimal_k}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.stats import multivariate_normal
import matplotlib.cm as cm
from matplotlib.patches import Ellipse

# Implementarea algoritmului EM
def init_params(X, n_clusters):
    """Iniializarea parametrilor"""
    n_samples, n_features = X.shape
    
    weights = np.ones(n_clusters) / n_clusters
    
    means = X[np.random.choice(n_samples, n_clusters, replace=False)]
    
    covs = [np.eye(n_features) for _ in range(n_clusters)]
    
    return weights, means, covs

def e_step(X, weights, means, covs):
    """Pasul E (Expectation): calcularea responsabilitilor"""
    n_samples, n_features = X.shape
    n_clusters = len(weights)
    
    resp = np.zeros((n_samples, n_clusters))
    
    for k in range(n_clusters):
        resp[:, k] = weights[k] * multivariate_normal.pdf(X, mean=means[k], cov=covs[k])
    
    resp_sum = resp.sum(axis=1, keepdims=True)
    resp = resp / resp_sum
    
    return resp

def m_step(X, resp):
    """Pasul M (Maximization): actualizarea parametrilor"""
    n_samples, n_features = X.shape
    n_clusters = resp.shape[1]
    
    nk = resp.sum(axis=0)
    weights = nk / n_samples
    
    means = np.dot(resp.T, X) / nk[:, np.newaxis]
    
    covs = []
    for k in range(n_clusters):
        diff = X - means[k]
        cov_k = np.dot(resp[:, k] * diff.T, diff) / nk[k]
        covs.append(cov_k)
    
    return weights, means, covs

def calculate_log_likelihood(X, weights, means, covs):
    """Calcularea log-verosimilitii cu stabilitate numeric mbuntit"""
    n_samples = X.shape[0]
    n_clusters = len(weights)
    
    log_likelihood = 0
    

    for i in range(n_samples):
        log_probs = np.zeros(n_clusters)
        for k in range(n_clusters):
            log_probs[k] = np.log(weights[k]) + multivariate_normal.logpdf(X[i], mean=means[k], cov=covs[k])
        
        max_log_prob = np.max(log_probs)
        log_likelihood += max_log_prob + np.log(np.sum(np.exp(log_probs - max_log_prob)))
    
    return log_likelihood

def plot_clusters(X, resp, means, covs, iteration):
    plt.figure(figsize=(10, 8))
    
    colors = cm.viridis(np.linspace(0, 1, len(means)))
    
    labels = np.argmax(resp, axis=1)
    for i, color in enumerate(colors):
        plt.scatter(X[labels == i, 0], X[labels == i, 1], color=color, alpha=0.6, s=30)
    
    plt.scatter(means[:, 0], means[:, 1], c='black', s=100, marker='X')
    
    for i in range(len(means)):
        eigenvalues, eigenvectors = np.linalg.eigh(covs[i])
        
        angle = np.degrees(np.arctan2(eigenvectors[1, 0], eigenvectors[0, 0]))
        
        width, height = 2 * np.sqrt(eigenvalues) * 2
        
        ellipse = Ellipse(xy=means[i], width=width, height=height, angle=angle, 
                          fill=False, color=colors[i], linewidth=2)
        plt.gca().add_patch(ellipse)
    
    plt.title(f'Clustere la iteraia {iteration}')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.grid(True, alpha=0.3)
    plt.show()

def em_algorithm(X, n_clusters=3, max_iter=100, tol=1e-4, verbose=True, plot_every=5):
    """Algoritmul EM corectat cu log-verosimilitate normalizat"""
    weights, means, covs = init_params(X, n_clusters)
    log_likelihoods = []

    for iteration in range(max_iter):
        resp = e_step(X, weights, means, covs)
        weights, means, covs = m_step(X, resp)

        log_likelihood = calculate_log_likelihood(X, weights, means, covs)
        log_likelihoods.append(log_likelihood)

        if verbose and (iteration + 1) % plot_every == 0:
            print(f"Iteraia {iteration + 1}")
            plot_clusters(X, resp, means, covs, iteration + 1)

        if iteration > 0 and abs(log_likelihoods[-1] - log_likelihoods[-2]) < tol:
            if verbose:
                print(f"\nConvergen atins la iteraia {iteration + 1}")
            break

    return weights, means, covs, resp, log_likelihoods

# Rularea algoritmului EM
weights, means, covs, resp, log_likelihoods = em_algorithm(data, n_clusters=3, max_iter=20, plot_every=4)

# Evaluarea rezultatelor
predicted_labels = np.argmax(resp, axis=1)

# Compararea cu etichetele adevrate (doar pentru vizualizare)
plt.figure(figsize=(15, 6))


plt.subplot(1, 2, 2)
plt.scatter(data[:, 0], data[:, 1], c=predicted_labels, cmap='viridis', alpha=0.6, s=30)
plt.title('Clustering cu EM')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Afiarea parametrilor finali
print("\nParametrii finali ai modelului:")
for i in range(len(weights)):
    print(f"Cluster {i+1}:")
    print(f"  Pondere: {weights[i]:.4f}")
    print(f"  Medie: {means[i]}")
    print(f"  Matricea de covarian:")
    print(f"{covs[i]}")
    print()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.stats import multivariate_normal
import matplotlib.cm as cm

# Generarea datelor
np.random.seed(101)
data, y = make_blobs(n_samples=288, n_features=2, centers=3, cluster_std=1.8, random_state=101)

# Vizualizarea datelor iniiale
plt.figure(figsize=(10, 6))
plt.scatter(data[:, 0], data[:, 1], c=y, cmap='viridis', alpha=0.7)
plt.title('Date generate cu make_blobs')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(alpha=0.3)
plt.show()

# Algoritm EM Simplificat
def em_clustering(X, k=3, max_iter=20, tol=1e-4):
    """
    Implementare simplificat a algoritmului Expectation-Maximization
    
    Args:
        X: Date de intrare (n_samples, n_features)
        k: Numrul de clustere
        max_iter: Numrul maxim de iteraii
        tol: Tolerana pentru convergen
        
    Returns:
        weights: Ponderile componentelor
        means: Mediile clusterelor
        covs: Matricile de covarian
        labels: Etichetele clusterelor
    """
    n_samples, n_features = X.shape
    
    # 1. Iniializare
    # Ponderi egale pentru toate clusterele
    weights = np.ones(k) / k
    
    # Alegerea aleatoare a centrelor iniiale
    idx = np.random.choice(n_samples, k, replace=False)
    means = X[idx]
    
    # Matrici de covarian iniiale
    covs = [np.eye(n_features) for _ in range(k)]
    
    # Pentru urmrirea convergenei
    log_likelihood_old = 0
    
    # Lista pentru stocarea log-verosimilitilor
    log_likelihoods = []
    
    # 2. Iteraii EM
    for iteration in range(max_iter):
        # Pasul E: Calcularea responsabilitilor
        resp = np.zeros((n_samples, k))
        
        for j in range(k):
            resp[:, j] = weights[j] * multivariate_normal.pdf(X, mean=means[j], cov=covs[j])
        
        # Normalizarea responsabilitilor
        resp = resp / resp.sum(axis=1, keepdims=True)
        
        # Pasul M: Actualizarea parametrilor
        
        # Numrul efectiv de puncte n fiecare cluster
        nk = resp.sum(axis=0)
        
        # Actualizarea ponderilor
        weights = nk / n_samples
        
        # Actualizarea mediilor
        means = np.dot(resp.T, X) / nk[:, np.newaxis]
        
        # Actualizarea matricilor de covarian
        covs = []
        for j in range(k):
            diff = X - means[j]
            cov_j = np.dot(resp[:, j] * diff.T, diff) / nk[j]
            covs.append(cov_j)
        
        # Calcularea log-verosimilitii
        log_likelihood = np.sum(np.log(np.sum(resp * resp / resp.sum(axis=1, keepdims=True), axis=1)))
        log_likelihoods.append(log_likelihood)
        
        # Verificarea convergenei
        if iteration > 0 and abs(log_likelihood - log_likelihood_old) < tol:
            print(f"Convergen atins dup {iteration+1} iteraii")
            break
            
        log_likelihood_old = log_likelihood
        
        # Afiarea progresului
        if (iteration + 1) % 5 == 0 or iteration == 0:
            print(f"Iteraia {iteration+1}, Log-likelihood: {log_likelihood:.4f}")
    
    # Calcularea etichetelor finale
    labels = np.argmax(resp, axis=1)
    
    # Afiarea rezultatelor
    plt.figure(figsize=(10, 6))
    plt.plot(log_likelihoods)
    plt.title('Evoluia log-likelihood')
    plt.xlabel('Iteraie')
    plt.ylabel('Log-likelihood')
    plt.grid(alpha=0.3)
    plt.show()
    
    return weights, means, covs, labels

# Rularea algoritmului EM
print("Aplicarea algoritmului EM...")
weights, means, covs, labels = em_clustering(data, k=3)

# Vizualizarea rezultatelor
plt.figure(figsize=(12, 5))

# Clustere originale
plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1], c=y, cmap='viridis', alpha=0.7)
plt.title('Clustere originale')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(alpha=0.3)

# Clustere detectate
plt.subplot(1, 2, 2)
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', alpha=0.7)
plt.scatter(means[:, 0], means[:, 1], c='red', s=100, marker='X')
plt.title('Clustere identificate cu EM')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Afiarea parametrilor finali
print("\nParametrii finali ai modelului:")
for i in range(len(weights)):
    print(f"Cluster {i+1}:")
    print(f"  Pondere: {weights[i]:.4f}")
    print(f"  Medie: {means[i]}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.stats import multivariate_normal
import matplotlib.cm as cm
from matplotlib.patches import Ellipse


# Implementarea algoritmului EM
def init_params(X, n_clusters):
    """Iniializarea parametrilor"""
    n_samples, n_features = X.shape
    weights = np.ones(n_clusters) / n_clusters
    means = X[np.random.choice(n_samples, n_clusters, replace=False)]
    covs = [np.eye(n_features) for _ in range(n_clusters)]
    return weights, means, covs

def e_step(X, weights, means, covs):
    """Pasul E (Expectation): calcularea responsabilitilor"""
    n_samples, n_clusters = X.shape[0], len(weights)
    resp = np.zeros((n_samples, n_clusters))
    
    for k in range(n_clusters):
        resp[:, k] = weights[k] * multivariate_normal.pdf(X, mean=means[k], cov=covs[k])
    
    resp_sum = resp.sum(axis=1, keepdims=True)
    resp = np.divide(resp, resp_sum, where=(resp_sum != 0))  # Evitm mprirea la zero
    
    return resp

def m_step(X, resp):
    """Pasul M (Maximization): actualizarea parametrilor"""
    n_samples, n_features = X.shape
    n_clusters = resp.shape[1]
    
    nk = resp.sum(axis=0)
    weights = nk / n_samples
    means = np.dot(resp.T, X) / nk[:, np.newaxis]
    
    covs = []
    for k in range(n_clusters):
        diff = X - means[k]
        cov_k = np.dot((resp[:, k] * diff.T), diff) / nk[k]
        cov_k += np.eye(n_features) * 1e-6  # Evitm singularitatea matricei
        covs.append(cov_k)
    
    return weights, means, covs

def calculate_log_likelihood(X, weights, means, covs):
    """Calcularea log-verosimilitii i normalizarea acesteia"""
    n_samples, n_clusters = X.shape[0], len(weights)
    log_likelihood = np.zeros(n_samples)
    
    for k in range(n_clusters):
        log_likelihood += weights[k] * multivariate_normal.pdf(X, mean=means[k], cov=covs[k])
    
    log_likelihood = np.log(log_likelihood + 1e-9)  # Evitm log(0)
    return log_likelihood.mean()  # Normalizm prin numrul de eantioane

def plot_clusters(X, resp, means, covs, iteration):
    """Vizualizarea clusterelor"""
    plt.figure(figsize=(10, 8))
    colors = cm.viridis(np.linspace(0, 1, len(means)))
    
    labels = np.argmax(resp, axis=1)
    for i, color in enumerate(colors):
        plt.scatter(X[labels == i, 0], X[labels == i, 1], color=color, alpha=0.6, s=30)
    
    plt.scatter(means[:, 0], means[:, 1], c='black', s=100, marker='X')
    
    for i in range(len(means)):
        eigenvalues, eigenvectors = np.linalg.eigh(covs[i])
        angle = np.degrees(np.arctan2(eigenvectors[1, 0], eigenvectors[0, 0]))
        width, height = 2 * np.sqrt(eigenvalues) * 2
        ellipse = Ellipse(xy=means[i], width=width, height=height, angle=angle, 
                          fill=False, color=colors[i], linewidth=2)
        plt.gca().add_patch(ellipse)
    
    plt.title(f'Clustere la iteraia {iteration}')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.grid(True, alpha=0.3)
    plt.show()

def em_algorithm(X, n_clusters=3, max_iter=100, tol=1e-4, verbose=True, plot_every=5):
    """Algoritmul EM corectat cu log-verosimilitate normalizat"""
    weights, means, covs = init_params(X, n_clusters)
    log_likelihoods = []

    for iteration in range(max_iter):
        resp = e_step(X, weights, means, covs)
        weights, means, covs = m_step(X, resp)

        log_likelihood = calculate_log_likelihood(X, weights, means, covs)
        log_likelihoods.append(log_likelihood)

        if verbose and (iteration + 1) % plot_every == 0:
            print(f"Iteraia {iteration + 1}")
            plot_clusters(X, resp, means, covs, iteration + 1)

        if iteration > 0 and abs(log_likelihoods[-1] - log_likelihoods[-2]) < tol:
            if verbose:
                print(f"\nConvergen atins la iteraia {iteration + 1}")
            break

    return weights, means, covs, resp, log_likelihoods

# Rularea algoritmului EM
weights, means, covs, resp, log_likelihoods = em_algorithm(data, n_clusters=3, max_iter=20, plot_every=4)

# Evaluarea rezultatelor
predicted_labels = np.argmax(resp, axis=1)

# Compararea cu etichetele adevrate (doar pentru vizualizare)
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 2)
plt.scatter(data[:, 0], data[:, 1], c=predicted_labels, cmap='viridis', alpha=0.6, s=30)
plt.title('Clustering cu EM')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Afiarea parametrilor finali
print("\nParametrii finali ai modelului:")
for i in range(len(weights)):
    print(f"Cluster {i+1}:")
    print(f"  Pondere: {weights[i]:.4f}")
    print(f"  Medie: {means[i]}")
    print(f"  Matricea de covarian:")
    print(f"{covs[i]}\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs

# Crearea datelor
data, y = make_blobs(n_samples=288, n_features=2, 
                      centers=3, cluster_std=1.8, random_state=101)

# Iniializare GMM
gmm = GaussianMixture(n_components=3, random_state=42, init_params='random', covariance_type='full')

# Funcie pentru desenarea elipselor de covarian
def plot_gaussian_ellipses(gmm, ax):
    for i in range(gmm.n_components):
        mean = gmm.means_[i]
        cov = gmm.covariances_[i]
        eigenvalues, eigenvectors = np.linalg.eigh(cov)
        angle = np.degrees(np.arctan2(eigenvectors[1, 0], eigenvectors[0, 0]))
        width, height = 2 * np.sqrt(eigenvalues)
        ellipse = patches.Ellipse(xy=(mean[0], mean[1]), width=width, height=height, 
                                  angle=angle, edgecolor='black', facecolor='none')
        ax.add_patch(ellipse)

# Lista pentru stocarea rezultatelor
iterations = 10
for i in range(iterations):
    gmm.fit(data)
    labels = gmm.predict(data)
    
    # Vizualizarea clusterelor la fiecare iteraie
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', alpha=0.6)
    ax.scatter(gmm.means_[:, 0], gmm.means_[:, 1], c='black', marker='x', s=100, label='Centroids')
    plot_gaussian_ellipses(gmm, ax)
    ax.set_title(f'Iteraia {i+1}, Log-verosimilitate: {gmm.lower_bound_:.4f}')
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.legend()
    plt.show()
    
    # Verificarea convergenei
    if gmm.converged_:
        print(f'Algoritmul a convergat n {i+1} iteraii.')
        break
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# Modele de amestec Gaussiene (Clustering probabilistic)
def ruleaza_gmm(data):
    print("\n--- Modele de amestec Gaussiene (Probabilist) ---")
    print("Concept: GMM presupune c datele sunt generate dintr-un amestec de distribuii Gaussiene")
    print("Se folosete algoritmul EM pentru a gsi parametrii acestora.")
    print("Pai (Expectaie-Maximizare):")
    print("1. Iniializarea parametrilor (medii, covariante, ponderi) pentru componentele Gaussiene")
    print("2. Pasul E: Calcularea probabilitilor fiecrui punct de a aparine fiecrei distribuii Gaussiene")
    print("3. Pasul M: Actualizarea parametrilor pe baza acestor probabiliti")
    print("4. Repetarea pailor 2-3 pn la convergen")
    
    # Crearea modelului GMM
    n_clusters = 3
    gmm = GaussianMixture(n_components=n_clusters, random_state=42, covariance_type='full')
    
    # Antrenarea modelului
    gmm.fit(data)
    
    # Obinerea etichetelor clusterelor
    etichete = gmm.predict(data)
    
    # Obinerea centrelor componentelor
    centre = gmm.means_
    
    # Calcularea scorului Silhouette
    silhouette = silhouette_score(data, etichete)
    print(f"\nScorul Silhouette: {silhouette:.3f}")
    
    # Vizualizarea rezultatelor folosind plt.scatter
    plt.figure(figsize=(8, 6))
    plt.scatter(data[:, 0], data[:, 1], c=etichete, cmap='viridis', alpha=0.6)
    plt.scatter(centre[:, 0], centre[:, 1], c='red', marker='x', s=100, label='Centroide')
    plt.title(f"Model de amestec Gaussian (componente={n_clusters})")
    plt.legend()
    plt.show()

    # Vizualizarea componentelor Gaussiene (elipsa de contur)
    plt.figure(figsize=(10, 8))
    plt.scatter(data[:, 0], data[:, 1], s=50, alpha=0.7)
    
    for i in range(n_clusters):
        covarianta = gmm.covariances_[i]
        medie = gmm.means_[i]
        
        # Obinerea valorilor proprii i vectorilor proprii
        v, w = np.linalg.eigh(covarianta)
        unghI = np.arctan2(w[0, 1], w[0, 0]) * 180 / np.pi
        v = 2.0 * np.sqrt(2.0) * np.sqrt(v)
        
        # Desenarea elipsei (contururi de 95%)
        ell = Ellipse(xy=(medie[0], medie[1]), width=v[0], height=v[1], angle=unghI,
                     edgecolor='black', facecolor='none', linewidth=2)
        
        plt.gca().add_patch(ell)
    
    plt.title('Vizualizarea componentelor GMM')
    plt.xlabel('Caracteristica 1')
    plt.ylabel('Caracteristica 2')
    plt.grid(True, alpha=0.3)
    plt.show()
    
    return etichete, centre

# Exemplu de apel al funciei (presupunnd c `data` este definit):
# etichete_gmm, centre_gmm = ruleaza_gmm(data)
ruleaza_gmm(data)
```
